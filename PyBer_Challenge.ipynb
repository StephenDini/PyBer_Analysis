# Pyber Challenge

### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Deliverable 1: Get a Summary DataFrame 

```python
# In Step 1, use the groupby() function to create a Series of data that has the type of city as the index, then apply the count() method to the "ride_id" column.
#  1. Get the total rides for each city type
ride_count = pyber_data_df.groupby(["type"]).count()["ride_id"]
urban_ride_count = ride_count["Urban"]
suburban_ride_count = ride_count["Suburban"]
rural_ride_count = ride_count["Rural"]
total_ride_count = urban_ride_count + suburban_ride_count + rural_ride_count
ride_count.rename("Total Rides")
ride_count.head() 
```

```python
# In Step 2, use the groupby() function to create a Series of data that has the type of city as the index, then apply the sum() method to the "driver_count" column.
# 2. Get the total drivers for each city type
driver_count = city_data_df.groupby(["type"])["driver_count"].sum()
driver_urban_count = driver_count["Urban"]
driver_suburban_count = driver_count["Suburban"]
driver_rural_count = driver_count["Rural"]
total_driver_count = driver_urban_count + driver_suburban_count + driver_rural_count
driver_count.head()
```

```python
# In Step 3, use the groupby() function to create a Series of data that has the type of city as the index, then apply the sum() method to the "fare" column.
#  3. Get the total amount of fares for each city type
fare_amount = pyber_data_df.groupby(["type"]).sum()["fare"]
urban_fare_amount = fare_amount["Urban"]
suburban_amount = fare_amount["Suburban"]
rural_amount = fare_amount["Rural"]
total_fare_amount = urban_fare_amount + suburban_amount + rural_amount 
fare_amount.head()
```

```python
#  4. Get the average fare per ride for each city type. 
# In Step 4, calculate the average fare per ride by city type by dividing the sum of all the fares by the total rides.
average_fare_ride = fare_amount / ride_count
average_fare_ride.head()
```

```python
# 5. Get the average fare per driver for each city type. 
# In Step 5, calculate the average fare per driver by city type by dividing the sum of all the fares by the total drivers.
average_fare_driver = fare_amount / driver_count
average_fare_driver
```

```python
#  6. Create a PyBer summary DataFrame. 
# In Step 6, create a PyBer summary DataFrame with all the data gathered from Steps 1-5, using the column names shown below:
# Adding a list of values with keys to create a new DataFrame.
pyber_summary_df = {"Total Rides":ride_count, 
                    "Total Drivers":driver_count, 
                    "Total Fares":fare_amount,
                    "Average Fare per Ride":average_fare_ride, 
                    "Average Fare per Driver":average_fare_driver}
pyber_summary_df = pd.concat(pyber_summary_df, axis=1)
pyber_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
```

```python
#  8. Format the columns.
pyber_summary_df["Total Rides"] = pyber_summary_df["Total Rides"].map("{:,d}".format)
pyber_summary_df["Total Drivers"] = pyber_summary_df["Total Drivers"].map("{:,d}".format)
pyber_summary_df["Total Fares"] = pyber_summary_df["Total Fares"].map("${:,.2f}".format)
pyber_summary_df["Average Fare per Ride"] = pyber_summary_df["Average Fare per Ride"].map("${:,.2f}".format)
pyber_summary_df["Average Fare per Driver"] = pyber_summary_df["Average Fare per Driver"].map("${:,.2f}".format)
pyber_summary_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame
# In Step 1, create a new DataFrame with multiple indices using the groupby() function on the "type" and "date" 
# columns of the pyber_data_df DataFrame, then apply the sum() method on the "fare" column to show the total fare 
# amount for each date.
pyber_type_date_df = pyber_data_df.groupby(["type","date"])
pyber_type_date_df.head()
pyber_summation = pyber_type_date_df["fare"].sum()
pyber_summation
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
# In Step 2, use the provided code snippet to reset the index. 
# This is needed to use the pivot() function in the next step (Step 3).
```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
pyber_summation = pyber_summation.reset_index()
pyber_summation
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
# In Step 3, use the pivot() function to convert the DataFrame from the previous step so that 
# the index is the "date," each column is a city "type," and the values are the "fare."
pyber_pivot = pyber_summation.pivot(index="date", columns="type", values="fare")
pyber_pivot
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
located_df = pyber_pivot['2019-01-01':'2019-04-28']
located_df
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
# In Step 5, use the provided code snippet to reset the index of the DataFrame from the previous step 
# (Step 4) to a datetime data type. This is necessary to use the resample() method in Step 7.
located_df.index = pd.to_datetime(located_df.index)
located_df
```

```python
# 7. Check that the datatype for the index is datetime using df.info()
located_df.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
located_df = located_df.resample("W").sum()
located_df
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')
plt.rcParams["figure.figsize"] = (12,3)
located_df.plot()
plt.xlabel("")
plt.title("Total Fare by City Type")
plt.ylabel("Fare($USD)")
plt.legend(loc="center")
plt.savefig("analysis/PyBer_fare_summary.png")
```
