```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot.
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.scatter(y_axis,x_axis, color="red")
plt.xlabel('Fare($)')
plt.ylabel("Date")
plt.title("PyBer Fare by Month")
plt.xlim(0,50)
plt.gca().invert_yaxis()
plt.legend(("Chicago","Chicago"))
plt.show()
```

```python
fig, ax = plt.subplots()
# x_axis =[x* 5 for x in x_axis]
y_axis = [y*5 for y in y_axis]
ax.scatter(y_axis, x_axis, s=y_axis, fc=(.4, .8, 1, .8), edgecolor=(0,0,0,1), linewidth=2)
ax.legend(("Boston","Boston"))
ax.invert_yaxis()
```

```python
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0.2, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%',
    shadow=True,
    startangle=95,
    counterclock=False)

plt.show()
fig, ax = plt.subplots()
ax.pie(y_axis,labels=x_axis)
plt.show()
```
